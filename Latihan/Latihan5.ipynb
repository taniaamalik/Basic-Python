```python
def jumlahvokal(string):
  jumVokal = 0
  string = string.lower()
  for character in "aiueo":
    jumVokal += string.count(character)
  return jumVokal
string = input("masukkan kata: ")

jumlahvokal(string)
```

```python
def count_vowel(string):
  num_vowels = 0
  for char in string:
    if char in "aiueoAIUEO":
      num_vowels += 1
  return num_vowels

def replicate(string):
  return count_vowels(string)*string

print("Jumlah huruf vokal adalah: ", count_vowels("Saya"))
print("String diduplikasi berdasarkan jumlah huruf vokal: ", replicate("Saya"))
```

```python
# or
def convert_date(string):
    data=string.split('-')
    tahun=data[0]
    bulan=int(data[1])
    tanggal=data[2]
    nama_bulan={1:"Januari",2:"Februari",3:"Maret",4:"April",5:"Mei",6:"Juni",7:"Juli",8:"Agustus",9:"September",10:"Oktober",11:"November",12:"Desember"}
    bulan=nama_bulan[bulan]
    return ' '.join([tanggal, bulan, tahun])
tgl=input("YYYY-MM-DD: ")
print(convert_date(tgl))
```

```python
panjang = int(input("masukkan nilai panjang: "))
lebar = int(input("masukkan nilai lebar: "))

def luaspersegi(panjang, lebar):
  return panjang*lebar

print("Panjang : ", panjang)
print("Lebar : ", lebar)

try:
  if (panjang <= 0 and lebar <= 0):
    raise ValueError("Panjang dan Lebar negatif")
  elif panjang <= 0:
    raise ValueError("Panjang itu bukan bilangan positif")
  elif lebar <= 0:
    raise ValueError("Lebar itu bukan bilangan positif")
  else:
    print("Hasil luas panjang dikali lebar :", luaspersegi(panjang, lebar))
except ValueError as ve:
  print(ve)
```

```python
# or
def luas_persegi(panjang, lebar):
    if panjang<0:
        raise ValueError("Panjang tidak boleh negatif")
    if lebar<0:
        raise ValueError("Lebar tidak boleh negatif")
    return panjang*lebar
panjang=float(input("masukkan panjang: "))
lebar=float(input("masukkan lebar: "))
try:
    luas=luas_persegi(panjang, lebar)
    print("Luas Persegi {}x{}={}".format(panjang, lebar, luas))
except ValueError as err:
    print("Kesalahan", err)
```

```python
def hitung_pangkat1(angka, pangkat):
    try:
        hasil = angka ** pangkat
        return hasil
    except ValueError:
        return "Inputan bukan berupa angka"
    except OverflowError:
        return "Pangkat terlalu besar"
    except ArithmeticError:
        return "Kesalahan aritmatika"
    except Exception:
        return "Kesalahan tidak diketahui"

c = float(input("Enter the first number: "))
d = float(input("Enter the second number: "))

hitung_pangkat1(c, d)
```

```python
# or
def hitung_pangkat(angka, pangkat):
    try:
        j=float(angka)
        i=float(pangkat)
        j=j**i
        return j
    except ValueError as e:
        print(f"{e}, {e.__class__}")
    except OverflowError as e:
        print(f"{e}, {e.__class__}")
    except ArithmeticError as e:
        print(f"{e}, {e.__class__}")
    except Exception as e:
        print(f"{e}, {e.__class__}")
angka=input("Masukkan bilangan desimal: ")
pangkat=input("Masukkan pangkat: ")
hitung_pangkat(angka, pangkat)
```
