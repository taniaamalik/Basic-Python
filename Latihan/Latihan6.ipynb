```python
from math import pi

class Lingkaran:

  def __init__(self,jari):
    self.jari=jari
    if jari<0:
      self.jari-0
      print("Nilai jari-jari harus lebih dari nol")

  def luas_lingkaran(self):
    return round((pi*(self.jari**2)),1)

  def keliling_lingkaran(self):
    return round((pi*self.jari*2),1)
    
  def hasil(self):
    print("Luas lingkaran dengan jari-jari " +str(self.jari)+ " adalah " + str(self.luas_lingkaran()) + " cm2")
    print("Luas lingkaran dengan jari-jari " +str(self.jari)+ " adalah " + str(self.keliling_lingkaran()) + " cm")

jari = int(input("Silahkan masukkan jari-jari lingkaran dalam cm: "))
operasi = Lingkaran(jari)
operasi.hasil()
```

```python
class Penjumlahan:
  def __init__(self):
    self.sum = []
      
  def get_sum(self):
    return self.sum

  def operasi_sum(self, aa, bb):
    self.sum.append(aa + bb)

aa = [3, 6, 9, 12, 15] 
bb = [100, 200, 300, 400, 500]
jmlh = Penjumlahan()

for i in range(len(aa)):
  jmlh.operasi_sum(aa[i], bb[i])

print(jmlh.get_sum())
```

```python
# or
class sumlist():
    def __init__(self, my_list):
        self.mylist=my_list
    def __add__ (self, other):
        new_list=[x+y for x,y in zip(self.mylist, other.mylist)]
        return sumlist(new_list)
    def __repr__(self):
        return str(self.mylist)
aa=sumlist([3,6,9,12,15])
bb=sumlist([100,200,300,400,500])
cc=aa+bb
print(cc)
```

```python
class Dictionary:
  def __init__(self):
    self.__dict={}

  def set_item(self,key,value):
    if key not in self.__dict.keys():
      self.__dict.update({key:value})
    else:
      print("key sudah ada")

  def del_item(self,key):
    if key in self.__dict.keys():
      del self.__dict[key]
    else:
      print("data tidak ditemukan")
  
  def get_item(self,key):
    return self.__dict[key]

  def __str__(self):
    text = ""
    for key, value in self.__dict.items():
      text += str(key)+ " "+str(value)+"\n"
    return text

object = Dictionary()
object.set_item(1, "Januari")
object.set_item(2, "Februari")
object.set_item(3, "Maret")
object.set_item(4, "April")
object.set_item(5, "Mei")
object.del_item(5)
print(object)
```

```python
class Kalkulator:

  def add(num1, num2):
    print("Hasil Jumlah", num1+num2)
  def substract(num1, num2): 
    print("Hasil Kurang", num1-num2)
  def multiply(num1, num2):
    print("Hasil Kali", num1*num2)
  def divide(num1, num2): 
    if num2==0:
      print("Pembagian dengan nol")
    else:
      print("Hasil Bagi", num1/num2)

num1 = int(input("Masukkan bilangan pertama: "))
num2 = int(input("Masukkan bilangan kedua: "))
print("nomor ke 1 :", num1)
print("nomor ke 2 :", num2)

Kalkulator.add(num1,num2)
Kalkulator.substract(num1,num2)
Kalkulator.multiply(num1,num2)
Kalkulator.divide(num1,num2)
```

```python
# or
class calculator:
    """Simple Calculator Class"""
    def add(self, x, y):
        return x+y
    def substract(self, x, y):
        return x-y
    def multiply(self, x, y):
        return x*y
    def divide(self, x, y):
        return x/y
my_cl=calculator()
x=int(input("masukkan nilai x: "))
y=int(input("masukkan nilai y: "))
print("hasil penjumlahan=", my_cl.add(x,y))
print("hasil pengurangan=", my_cl.substract(x,y))
print("hasil perkalian=", my_cl.multiply(x,y))
print("hasil pembagian=", my_cl.divide(x,y))
```
