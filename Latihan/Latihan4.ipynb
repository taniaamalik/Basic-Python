Buatlah perintah sederhana untuk menampilkan nilai “pi” dari import numpy, import math, dan import scipy!

```python
import numpy
import math
import scipy

print(numpy.pi)
print(math.pi)
print(scipy.pi)
```

Buatlah perintah sederhana untuk membuat aliasing import math, misal dengan alias “mm”, lalu tampilan hasil nilai e dengan alias tersebut!

```python
import math as mm

print(mm.e)
```

Buatlah perintah sederhana untuk membuat double aliasing misal from math import pi, misal dengan alias “PI”, lalu import sin, misal dengan alias “sine”. Lalu tampilan hasil print nilai “sine(PI/2)” dengan alias tersebut!

```python
from math import pi as PI, sin as sine

print(sine(PI/2))
```
