```python
def hitungList(inList,pengali=100):
    hasil=0
    for i in range(len(inList)):
        hasil += inList[i]*pengali
    return hasil

inList=[1,2,3]
pengali=2
print(hitungList(inList,pengali))
```

```python
# or
def jumlah_kali(a, b=100):
    total = 0
    for x in a:
        total += x
    total *= b
    return total
print(jumlah_kali((1,2,3),2))
```

Hasil analisis : Setelah dilakukan perbaikan nama fungsi, pada saat memasukan Nilai 1 > Nilai 2 akan terjadi error, karena fungsi akan terus repetisi sampai x==y dan akan selalu masuk ke else dimana akan selalu dikurangi 1, oleh karena itu diperlukan penambahan kondisi baru yaitu apabila Nilai 1 > Nilai 2 maka akan mengembalikan fungsi dengan parameter (x-1,y)

```python
def fungsi(x,y):
  if x == y:
    return x
  else:
    return fungsi(x, y-1)

a = int(input("masukkan nilai pertama: "))
b = int(input("masukkan nilai kedua: "))

print(fungsi(a,b))
```

```python
def fungsi(x,y):
  if x == y:
    return x
  elif x > y:
    return fungsi(x-1, y)
  else:
    return fungsi(x, y-1)

a = int(input("masukkan nilai pertama: "))
b = int(input("masukkan nilai kedua: "))

print(fungsi(a,b))
```

```python
cek_tuple = ("FGA", 1, 2, "FGA", 3, 4, 5, "FGA")
# cek_tuple = (1,2,3,4)
temp = set() # dengan bantuan method set >> digunakan untuk menyimpan beberapa item pada satu variable 
distinct = True
for i in cek_tuple:
  if i in temp:
    distinct = False
    break
  temp.add(i)
print("Apakah isi dari sebuah tuple distinct?", distinct)
```
