```python
angka_ganjil = []
for i in range(100):
    if i % 2 != 0:
        angka_ganjil.append(i)

print("Angka ganjil 1 sampai 100:")
print(angka_ganjil)

angka_kuadrat = []

for number in angka_ganjil:
    kuadrat = number ** 2
    angka_kuadrat.append(kuadrat)

print("\nKuadrat dari angka ganjil:")
print(angka_kuadrat)

total = 0
for number in angka_kuadrat:
    total = total + number

print("\nJumlah keseluruhan angka ganjil yang dikuadratkan:")
print(total)
print()

jumlah_angka = len(angka_kuadrat)
rata_rata = total/jumlah_angka

print("Rata-rata angka ganjil kuadrat:")
print("Jumlah angka ganjil ada {} angka".format(jumlah_angka))
print("Rata-rata: {}".format(rata_rata))
```

```python
in_lst = [1,3,7,'a',3,'a','b','c']
out_lst = []
match = False

for i in range(len(in_lst)):
  if not out_lst:
    out_lst.append(in_lst[i])
  else:
    for j in out_lst:
      if in_lst[i] == j:
        match = True
        continue
    if match != True:
      out_lst.append(in_lst[i])
    match = False

print(out_lst)
```
